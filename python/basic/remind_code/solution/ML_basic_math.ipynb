# 미분

### 함수 f(w)의 w에 대한 미분
$$
\frac{df(w)}{dw},
\frac{d}{dw}f(w), 
f'(w)
$$

### 함수 $f(w)=w^2$의 w에 대한 미분
$$
\frac{d}{dw}f(w) = \frac{d}{dw}w^2 = 2w
$$

### 함수 $f(w)=w^n$의 w에 대한 미분
$$
\frac{d}{dw}f(w) = \frac{d}{dw}w^n = nw^{n-1}
$$

```python
f(x) = x**2     x대한 식(1)
dx_f(x) = 2x   .. 식(1)에 대해서 x에 대한 1차 미분

f(x,w) = x**2+w**2     x대한 식(2)
dx_f(x,w) = 2x .. 식(2)에 대해서 x에 대한 1차 미분
dw_f(x,w) = 2w .. 식(2)에 대해서 w에 대한 1차 미분
```

### 행렬
####  $$행렬 A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$$


#### $$ 역행렬 AA^{-1}=A^{-1}A=I$$
$$역행렬이란, 곱하면 단위행렬 I가 되는 행렬$$
$$A^{-1} = \frac{1}{ad-bc}\begin{bmatrix} d & -b \\ -c & a \end{bmatrix}$$

```python
import numpy as np
A = np.array([[1,2],[3,4]])
A_Inv = np.linalg.inv(A)
print('A_Inv = ', A_Inv)
I = A_Inv.dot(A)
I = np.round(I,2)
print('I = ', I)
I = A.dot(A_Inv)
I = np.round(I,2)
print('I = ', I)

I = np.identity(10)
print('I = ', I)
```

```python
import numpy as np
import matplotlib.pyplot as plt

import copy
def binary_step(_x):
    x = copy.deepcopy(_x)
    for i,v in enumerate(x): 
        if v>0 : x[i]=1.0
        if v<0 : x[i]=0.0
    return x
def sigmoid_function(x):
    z = (1/(1 + np.exp(-x)))
    return z    
def tanh_function(x):
    z = (2/(1 + np.exp(-2*x))) -1
    return z
def relu_function(_x):
    x = copy.deepcopy(_x)
    for i,v in enumerate(x): 
        if v>0 : x[i]=x[i]
        if v<0 : x[i]=0.0
    return x

def leaky_relu_function(_x,a=0.051):
    data = [max(a*value,value) for value in _x]
    return np.array(data, dtype=float)
def elu_function(_x, a):
    x = copy.deepcopy(_x)
    for i,v in enumerate(x): 
        if v>0 : x[i]=x[i]
        if v<0 : x[i]=a*(np.exp(x[i])-1)
    return x

def swish_function(_x):
    x = copy.deepcopy(_x)
    return x/(1-np.exp(-x))
def softmax_function(_x):
    x = copy.deepcopy(_x)
    z = np.exp(x)
    z_ = z/z.sum()
    return z_

x = np.linspace(-10,10,100)
plt.figure(figsize=(5,5))
#plt.plot(x,binary_step(x),'gray', linestyle='--', linewidth=1, label='binary_step')
#plt.plot(x,sigmoid_function(x),'black', linewidth=1, label='sigmoid')
#plt.plot(x,tanh_function(x),'green', linewidth=1, label='tanh')
#plt.plot(x,relu_function(x),'orange', linewidth=1, label='relu')
#plt.plot(x,leaky_relu_function(x),'red', linewidth=1, label='leaky_relu')
#plt.plot(x,elu_function(x,1),'blue', linestyle='-.', linewidth=1, label='elu')
#plt.plot(x,swish_function(x),'cornflowerblue', linewidth=1, label='swish')
plt.plot(x,softmax_function(x),'violet', linewidth=1, label='softmax')

plt.xlim(-10,10)
plt.ylim(-5,5)
plt.grid(True)
plt.legend(loc='upper left')
```
